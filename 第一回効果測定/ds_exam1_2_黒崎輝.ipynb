# １．以下の指示に従いプログラムを各セルに実装してください。

➀ １以上、１０００未満の一様分布の乱数を１０個発生させて、２行５列の行列を生する

② １００以上、２０００未満の一様分布の乱数を１５個発生させて、３行５列の行列を生成する

③ ➀②で生成した行列をそれぞれ表示する

以上、配点　５点

```python
# 乱数シードの設定
np.random.seed(0)

# 以下実装
# ①
matrix1 = np.random.randint(1, 1000, (2, 5))
# ②
matrix2 = np.random.randint(100, 2000, (3, 5))
# ③
print("matrix1 = ")
print(matrix1)

print("matrix2 = ")
print(matrix2)
```

### ▶実行結果
kouka_1_1.PNG

➃ ➀②で生成した行列を結合して、５行５列の行列を生成する

➄ ➃で生成した行列を表示する

以上、配点　５点

```python
# 以下実装
# ④
ans4 = np.concatenate([matrix1, matrix2], axis = 0)
# ⑤
print("matrix1 と matrix2 の行方向への結合 =")
print(ans4)
```

### ▶実行結果
kouka_1_2.PNG

➅　④で作成した各列ごとの最大値、最小値、合計、平均を表示

以上、配点　５点

```python
# 以下実装
# ⑥
len = ans4.shape
for i in range(len[1]):
  print(str(i) + "列目")
  print("最大値 = ", np.max(ans4[:, i]))
  print("最小値 = ", np.min(ans4[:, i]))
  print("合計値 = ", np.sum(ans4[:, i]))
  print("平均値 = ", np.mean(ans4[:, i]))
```

### ▶実行結果

⑦ ➃で作成した５行５列の行列を下図のように、赤、青、緑の３つの部分に分割してそれぞれを表示

以上、配点　10点

```python
# 以下実装⑦
print("０～１列目")
print(ans4[:, 0:2])
print("２列目")
print(ans4[:, 2:3])
print("３～４列目")
print(ans4[:, 3:])
```
