**スライスの自習問題**

必ずＰ９５～を自分で確認しながら学習を進めてください

```python
# データを作成
import numpy as np
test = np.arange(1,101)
test = test.reshape(10,10)
test
```

```python
# 例1：testの 0 列目のデータだけ取り出しなさい
a = test[:,0]
a
```

```python
# 例2：testの 0 行目のデータだけ取り出しなさい
a = test[0,:]
a
```

```python
# ①：test の5行目のデータだけ取り出しなさい
a = test[5]
a
```

```python
# ②：testの8列目のデータだけ取り出しなさい
a = test[:, 8]
a
```

```python
# ③：testの最後の列だけをマイナスの値を使って取り出しなさい
a = test[-1]
a
```

```python
# ④：testの0行目から3行目までの値を取り出しなさい
# (0行目、1行目、2行目、3行目の合計4行が対象)
a = test[0:4]
a
```

```python
# ⑤：testの 3列目から4列目までの値を取り出しなさい
# 3列目、4列目の2列が対象
a = test[:, 3:5]
a
```

```python
# ⑥：testの 0行目から3行目かつ、3列目から4列目までの値を取り出しなさい
a = test[0:4, 3:5]
a
```
