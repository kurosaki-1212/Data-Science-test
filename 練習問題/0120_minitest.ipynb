```python
import numpy as np

line = '-------------------'
print(line)

# 乱数シード値は以降すべて10に設定すること
np.random.seed(10)

# 1以上、100未満の一様分布の乱数を20個発生させて、4行5列の行列を生成し変数m1に代入する
m1 = np.random.randint(1,100,(4, 5))
# -100 以上、100未満の一様分布の乱数を20個発生させて、4行5列の行列を生成し変数m2に代入する
np.random.seed(10)
m2 = np.random.randint(-100,100,(4, 5))

# m1、m2をそれぞれprintで表示する
print(m1)
print(line)

print(m2)
print(line)
# concatenate関数を使って、m1とm2を列方向に連結し、m3に代入する
m3 = np.concatenate([m1,m2], axis=1)

# m3をprintで表示する
print(m3)
print(line)

# 出力例の図の青い破線部分をスライスで取り出して出力例１のとおり表示してください
print('出力例１')
print(m3[2,:])
print(line)

# 出力例の図の赤い枠線部分をスライスで取り出し表示例の２のとおり表示してください
print('出力例２')
print(m3[2:,-3:])
```
