「data.csv」を読み込んで、pandas の機能を使って、「商品」列を One Hot Encoding し、結果を表示してください。


```python
# 必要なライブラリを読み込む
import pandas as pd
```

```python
# 「data.csv」を読み込んで、df に データを格納する
df = pd.read_csv("data.csv", encoding = "shift-jis")

# df の内容を表示する
df
```

```python
# df から 商品列だけを抜き出し、df2 に代入する
# DataFrame['カラム名']で取得するとSeries型、DataFrame[['カラム名']]で取得するとDataFrame型です！！！！
df2 = df[["商品"]]

# df2の内容を表示
df2
```

```python
#  df2 の中身を One Hot Encoding し、newdf に代入する
newdf = pd.get_dummies(df2, prefix = "商品")

# newdfの中身を表示
newdf
```
