「0203_sushi.jpg」を読み込んで、以下のコメント通りにプログラムを実装してください。

```python
# cv2のインポート
import cv2

#「0203_sushi.jpg」という画像データを読み込んでimg という変数で使えるようにする
img = cv2.imread("0203_sushi.jpg")

# google.colab環境上で画像を表示する
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
# データ形状を表示（高さ、幅、チャネル）
img.shape
```

```python
# 80% のスケールにする(80%に縮小する)
# 80%のサイズに大きさを変更し、img_resized という変数に代入する
img_resized = cv2.resize(img, None, fx = 0.8, fy = 0.8)

# 変更した画像を表示する
cv2_imshow(img_resized)
```

```python
# サイズ変更した img_resized のデータ形状を表示（高さ、幅、チャネルを表示）
img_resized.shape
```

```python
# 画像を右に90度回転し、img_rotated に代入する
img_rotated = cv2.rotate(img_resized, cv2.ROTATE_90_CLOCKWISE)

# 回転させた画像（img_rotated）を表示する
cv2_imshow(img_rotated)
```

```python
# 回転した画像（img_rotated）の画像のサイズを再確認する
img_rotated.shape
```

```python
# 回転した画像（img_rotated）から、上から4切れ分の寿司の画像のみ四角形に切り取り、sushi4 という変数に代入する
sushi4 = img_rotated[10:240, 80:190, :]

# 切り出した画像 sushi4 を表示する
cv2_imshow(sushi4)
```

```python
# 切り出した画像 sushi4 の画像データを保存する
# 保存するファイル名は、「0203_ds_氏名.jpg」
cv2.imwrite("0203_ds_黒崎輝.jpg", sushi4)
```
