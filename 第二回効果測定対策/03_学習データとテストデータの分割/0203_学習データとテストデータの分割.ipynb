学習データ、テストデータに分割する練習

```python
# pandas のインポート
import pandas as pd

# yachin.csvを読み込み、df という変数に格納する。
df = pd.read_csv("yachin.csv")
df
```

```python
# 説明変数の列（=hirosa,eki,chikunenの3つの列）のみ取り出して、Xに代入する
X = df.iloc[:, :-1]
# 目的変数の列（=yachin列）のみ取り出して、yに代入する
y = df.iloc[:, -1]

# データ分割（学習データとテストデータ）のための train_test_split 関数のインポート
from sklearn.model_selection import train_test_split

# 学習データとテストデータに分け、X_train, X_test, y_train, y_testという変数に代入する
# 分割は 60%を学習データにし、40%をテストデータにする
# random_state=0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 0)
```
