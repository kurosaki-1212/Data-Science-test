「Harry_Potter_Movies_Characters.csv」を読み込んで、以下のコメントどおりのプログラムを実装してください。

```python
# pandasのインポート
import pandas as pd
```

```python
# 「Harry_Potter_Movies_Characters.csv」を読み込み、df という変数に格納する
df = pd.read_csv("Harry_Potter_Movies_Characters.csv")
```

```python
# dfの先頭から10行を表示する
df.head(10)
```

```python
# 例：df2 に 「Character Name」だけを列名を指定して代入する。ただし、df2にはSeries型で代入すること
df2 = df['Character Name']
df2
```

```python
# df3 に 「Character Name」だけを取得し、代入する。ただし、locをつかってデータを取得し、Series型で取得しdf3に代入すること
df3 = df.loc[:, "Character Name"]

df3
```

```python
# df4 に 「Character Name」だけを取得し、代入する。ただし、ilocをつかってデータを取得し、Series型で取得しdf4に代入すること
df4 = df.iloc[:, 1]
df4
```

```python
# df5 に 「Character Name」だけを列名を指定して代入する。ただし、df5 にはDataFrame型で代入すること
df5 = df[["Character Name"]]

df5
```

```python
# df6 に 「Character Name」と「House」を取得し、代入する。ただし、locをつかってデータを取得すること
# ヒント：隣り合わないデータは、stepを指定して取り出すことも可能
# DataFrame.loc[start:stop:step ,start:stop:step]
df6 = df.loc[:, "Character Name":"House":3]
df6 = df.loc[:, ["Character Name", "House"]]
df6 = df.iloc[:, 1:5:3]
df6
```

```python
# df7 に 「Character Name」と「House」と「Patronus」を取得し、代入する。
# 取得する列名のリストを作成してから、対象となるデータをdf7に格納すること
target_col=['Character Name','House','Patronus']
df7 = df[target_col]
df7
```

```python
# dfの内容を「Character Name」の昇順に並び替えて、最初の10行を表示してください。
df.sort_values('Character Name', ascending=True).head(10)
```

```python
# dfの内容を「Character Name」の降順に並び替えて、最初の15行を表示してください。
df.sort_values('Character Name', ascending=False).head(15)
```

```python
# dfの内容を「Patronus」「Character Name」の昇順に並び替えて、最初の20行を表示してください。
# ヒント：複数のキーで並び替えをしたい場合は、引数のbyを記載せず、並べ替えたい列名のリストをセットする
li = ["Patronus", "Character Name"]
dfli = df.sort_values(li, ascending = True)
dfli.head(20)
```
