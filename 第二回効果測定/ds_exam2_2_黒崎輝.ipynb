```python
# cv2のインポート
import cv2

#「ds_koukasokutei2-2.jpg」を読み込む。
img = cv2.imread("ds_koukasokutei2-2.jpg")

# 実行環境上で画像を表示する
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
# データ形状を表示（高さ、幅、チャネル）
print(img.shape)
```

```python
# 25% のスケールにする(25%に縮小する)
# 25%のサイズに大きさを変更し、img_resized という変数に代入する
img_resized = cv2.resize(img, None, fx = 0.25, fy = 0.25)

# 変更した画像を表示する
cv2_imshow(img_resized)
```

```python
# 画像を左に90度回転し、img_rotated に代入する
img_rotated = cv2.rotate(img_resized, cv2.ROTATE_90_COUNTERCLOCKWISE)

# 回転させた画像（img_rotated）を表示する
cv2_imshow(img_rotated)
```

```python
# 回転した画像（img_rotated）から、料理部分だけを四角形に切り取る
ramen = img_rotated[25:, 70:280, :]

# 切り出した画像 ramen を表示する
cv2_imshow(ramen)
```

```python
# ramenの画像データを 15×15 で平滑化する。
ramen_smoothed = cv2.blur(ramen,(15,15))
cv2_imshow(ramen_smoothed)
```

```python
# 任意の位置に、図形を描画する。
x0, y0 = 85, 75
x1, y1 = 130, 120
color = (0, 0, 255)
thickness = 3

ramen_rect = cv2.rectangle(ramen_smoothed, (x0, y0), (x1, y1), color, thickness)
cv2_imshow(ramen_rect)
```

```python
# 切り出した画像 sushi4 の画像データを保存する
cv2.imwrite("ds_koukasokutei2-2黒崎輝.jpg", ramen_rect)
```

```python
th, ramen_th= cv2.threshold(ramen, 60, 255, cv2.THRESH_BINARY)
cv2_imshow(ramen_th)
```

```python
# 任意の位置に、図形を描画する。
x0, y0 = 105, 100
color = (0, 0, 255)
thickness = 3

ramen_cicle = cv2.circle(ramen_th, (x0, y0), 30, color, thickness)
cv2_imshow(ramen_cicle)
```
