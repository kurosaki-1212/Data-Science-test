#第２回効果測定 問1

キノコのデータ群である「mushroom-test.csv」を読み込み、食べられるキノコか毒キノコかを判別するモデルを作成します。 それぞれのコメント欄をよく読み、適切なソースコードを記載して、ipynbファイルを提出してください。

提出するファイル名は、「ds_exam2-1_氏名.ipynb」とすること。


```python
# pandasのライブラリ
import pandas as pd
# 機械学習ライブラリ
import sklearn
import numpy as np
```

```python
# データの読み込み
mushroom = pd.read_csv("mushroom-test.csv", encoding="SHIFT-JIS")

# 先頭10行を表示する　(1)
mushroom.head(10)
```

```python
# mashroomデータの行数、列数を出力例のとおり表示する(2)
print("データの形式:" + str(mushroom.shape))

# mashroomデータ内の欠損値の数を出力例のとおり表示する(3)
print("データの欠損地:" + str(mushroom.isnull().sum().sum()))
```

**【解説】**

このキノコデータの目的変数は「class」です。これがpの場合は毒キノコ、eの場合は食用であることを示します。
それ以外のデータは説明変数です。

```python
# 説明変数をすべて One-Hot Encoding し、mushroom_dummyという新しいDataFrameに格納する(4)
mushroom_dummy = pd.get_dummies(mushroom.loc[:, ["傘の裏の色", "ひだ", "匂い", "傘の色"]])

# mushroom_dummyを表示
mushroom_dummy.head()
```

```python
# 目的変数をフラグ化（0/1化）して、mushroom_dummyに「flg」という新しい列として追加する
# 「class」列が「p」の場合、1、それ以外の場合は 0 として「flg」という列にし、目的変数とする
mushroom_dummy["flg"] = mushroom["class"].map(lambda x: 1 if x =="p" else 0)

# データが存在しているか確認
mushroom_dummy.head()
```

```python
# SVMのライブラリの読み込み
from sklearn.svm import LinearSVC
# 学習データとテストデータを分けるライブラリの読み込み
from sklearn.model_selection import train_test_split

# yに目的変数の列を代入する(5)
y = mushroom_dummy["flg"]

# Xに説明変数の列を代入する(6)
X = mushroom_dummy.iloc[:, :-1]

# 学習データとテストデータに分割する(7)
# test_size = 0.2 ,random_state = 0 とすること
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# モデルの初期化と学習を行う(8)
model = LinearSVC()
model.fit(X_train, y_train)

# 学習データの正解率の出力(9)
print("正解率(train):{:.3f}".format(model.score(X_train, y_train)))
```

```python
# テストデータで予測を行う(10)
result = model.predict(X_test)

# テストデータの結果を表示する
np.set_printoptions(threshold=np.inf)
print(result)
```

```python
# テストデータの正解率の出力
print("正解率(test)):{:.3f}".format(model.score(X_test, y_test)))
```
