Numpy には、np.max以外にも、平均や合計を求める関数が存在する。

最大値、最小値、平均、合計の求め方、使い方を確認しておきましょう。
ヒント：テキストP132等

※ 平均は2種類の方法があるが、当面はmeanを使用してよい
参考：https://deepage.net/features/numpy-average.html

```python
# 0行目の平均値を求めて表示する
print("0行目の平均値は(np.averageを使用)", np.average(ans6[0]))
print("0行目の平均値は(np.meanを使用)", np.mean(ans6[0]))

# 3列目の合計値を求めて表示する
print("3列目の合計値は(np.maxを使用)", np.sum(ans6[:, 3]))

# 4行目のの最小値を求めて表示する
print("4行目のの最小値は(np.minを使用)", np.min(ans6[4]))

# すべてのデータの中央値を求める
print("すべてのデータの中央値は(medianを使用を使用)", np.median(ans6))
```
