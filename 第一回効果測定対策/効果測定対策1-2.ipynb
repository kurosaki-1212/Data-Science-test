### ➆ ２０１０年１月～２０２１年１月までの各都市の最高気温、最低気温を表示
### 以上、配点　５点

```python
# 以下実装
print("東京の２０１０年１月～２０２１年１月までの最高気温 = {:.3f}℃".format(kisho_df_1.iloc[:, 2].max()))
print("東京の２０１０年１月～２０２１年１月までの最低気温 = {:.3f}℃".format(kisho_df_1.iloc[:, 3].min()))
print("東京の２０１０年１月～２０２１年１月までの最高気温 = {:.3f}℃".format(kisho_df_1.iloc[:, 5].max()))
print("東京の２０１０年１月～２０２１年１月までの最低気温 = {:.3f}℃".format(kisho_df_1.iloc[:, 6].min()))
print("東京の２０１０年１月～２０２１年１月までの最高気温 = {:.3f}℃".format(kisho_df_1.iloc[:, 8].max()))
print("東京の２０１０年１月～２０２１年１月までの最低気温 = {:.3f}℃".format(kisho_df_1.iloc[:, 9].min()))
```

### ⑫ ５都市の平均気温の箱ひげ図を表示
### 以上、配点　１０点

```python
# 以下実装
import matplotlib.pyplot as plt

import matplotlib.style

matplotlib.style.use('ggplot')

tokyodf = kisho_df_3.iloc[:, 1]
osakadf = kisho_df_3.iloc[:, 4]
hukuokadf = kisho_df_3.iloc[:, 7]
aomoridf = kisho_df_3.iloc[:, 10]
asahigawadf = kisho_df_3.iloc[:, 13]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

# グラフタイトル
ax.set_title('heikin kion')

# 表示ラベルを設定
labelsdata = ['Tokyo', "Osaka", "Hukuoka", "Aomori", "Asahigawa"]

# 箱ひげ図を描写
ax.boxplot((tokyodf,osakadf, hukuokadf, aomoridf, asahigawadf), labels=labelsdata)

# 表示
plt.show()
```

### ▶実行結果
kouka_1_16.png

### ⑬ 東京の平均気温と他４都市の平均気温との相関係数を表示
### 以上、配点　１０点

```python
# 以下実装

print("東京と大阪の平均気温の相関係数 = ", sp.stats.pearsonr(kisho_df_3.iloc[:, 1], kisho_df_3.iloc[:, 4]))
print("東京と福岡の平均気温の相関係数 = ", sp.stats.pearsonr(kisho_df_3.iloc[:, 1], kisho_df_3.iloc[:, 7]))
print("東京と青森の平均気温の相関係数 = ", sp.stats.pearsonr(kisho_df_3.iloc[:, 1], kisho_df_3.iloc[:, 10]))
print("東京と旭川の平均気温の相関係数 = ", sp.stats.pearsonr(kisho_df_3.iloc[:, 1], kisho_df_3.iloc[:, 13]))
```
