**② 10以上、100未満の一様分布の乱数を15個発生させて、3行5列の行列を生成する**

```python
np.random.seed(20)
matrix1 = np.random.randint(10, 100, (3, 5))
```

**④ 1以上、800未満の一様分布の乱数を20個発生させて、4行5列の行列を生成させ、表示例のように表示する。**

データは matrix2 という変数に格納すること。

▼表示例

```python
np.random.seed(20)
matrix2 = np.random.randint(1, 800, (4, 5))
print("matrix", matrix2)
```

**⑤　②と④で生成した行列を結合して、7行5列の行列を生成し画面に表示する。**
ヒント：テキストP112

```python
ans5 = np.concatenate([matrix1, matrix2], axis=0)
print(ans5)
```

**⑥ ⑤で作成したNumpy配列の転置を行い、表示する**

ヒント：テキストP116

↓実行例

1_1-5.png

```python
ans6 = ans5.T
print(ans6)
```

**⑦ ⑥で転置したデータの列ごとに最大値をを求め、以下の実行例のとおり表示する**

↓実行例

1_1-7.png


```python
for i in range(ans6.shape[1]):
  print(i, "列目の最大値は", np.max(ans6[:, i]))
```

Numpy には、np.max以外にも、平均や合計を求める関数が存在する。

最大値、最小値、平均、合計の求め方、使い方を確認しておきましょう。
ヒント：テキストP132等

※ 平均は2種類の方法があるが、当面はmeanを使用してよい
参考：https://deepage.net/features/numpy-average.html

```python
# 0行目の平均値を求めて表示する
print('0行目の平均値は(np.averageを使用)' ,np.average(ans6[0,:]))
print('0行目の平均値は(np.mean)',np.mean(ans6[0,:]))

# 3列目の合計値を求めて表示する
print("3行目の合計値は(np.sum)", np.sum(ans6[:, 3]))

# 4行目の最小値を求めて表示する
print("4行目の最小値は(np.min)", np.min(ans6[4]))

# すべてのデータの中央値を求める
print("全てのデータの中央値は(np.median)", np.median(ans6[:]))
```
